```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv(r"D:\uni dina\PSM2\_CODING\DATASET\preprocessed_data_v2.csv")
```

```python
df
```

```python
# Check for NaN values in each column
nan_count = df.isna().sum()

# Print the columns with NaN values and their respective counts
print("Columns with NaN values:")
print(nan_count[nan_count > 0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import model_selection, naive_bayes
from sklearn.metrics import classification_report, accuracy_score
```

```python
Train_X, Test_X, Train_Y, Test_Y = model_selection.train_test_split(df['tweet_content'],df['label'],test_size=0.3)
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(df['tweet_content'])
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
print(Tfidf_vect.vocabulary_)
```

```python
print(Train_X_Tfidf)
```

```python
# fit the training dataset on the NB classifier
Naive = naive_bayes.MultinomialNB()
Naive.fit(Train_X_Tfidf,Train_Y)

# Predict the labels on validation dataset
predictions_NB = Naive.predict(Test_X_Tfidf)

# Classification report
print("Classification Report:")
print(classification_report(Test_Y, predictions_NB))

# Use accuracy_score function to get the accuracy
print("Naive Bayes Accuracy Score -> ",accuracy_score(predictions_NB, Test_Y)*100)
```
