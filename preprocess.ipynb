```python
import re

# Define regex patterns for replacements
patterns = [
    (r"https?:\/\/\S+", ""),            # Remove URLs
    (r"pic\.twitter\.com\/\S+", ""),    # Remove pic.twitter.com URLs
    (r"@[A-Za-z0-9]+", ""),              # Remove mentions
    (r"><", " "),                        # Replace "<>" with space
    (r"[^a-zA-Z0-9\s]", ""),            # Remove non-alphanumeric characters except spaces
    (r"\b\d+\b", ""),                   # Remove numbers
    (r"\d", ""),                        # Remove all numerical characters
]

# Apply regex replacements to 'tweet_content' column
for pattern, replacement in patterns:
    corpus['tweet_content'] = corpus['tweet_content'].apply(lambda x: re.sub(pattern, replacement, x))

# Display the modified DataFrame
print(corpus)
```

```python
import nltk
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords corpus if not already downloaded
nltk.download('stopwords')

# Load English stopwords
stop_words = set(stopwords.words('english'))

# Function to remove stopwords from a text and convert to lowercase
def remove_stopwords(text):
    
    # Convert text to lowercase
    text = text.lower()
    
    # Tokenize the text into words
    word_tokens = word_tokenize(text)
    
    # Filter out stopwords from the word tokens
    filtered_tokens = [w for w in word_tokens if not w.lower() in stop_words]
    
    # Join the filtered tokens back into a single string
    return ' '.join(filtered_tokens)

# Apply the remove_stopwords function to 'tweet_content' column
corpus['tweet_content'] = corpus['tweet_content'].apply(remove_stopwords)

# Display the modified DataFrame
print(corpus)
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Define a function to stem words in a text
def stem_words(text):
    # Tokenize the text into words
    word_tokens = word_tokenize(text)
    # Stem each word in the tokenized text
    stems = [stemmer.stem(word) for word in word_tokens]
    # Join the stemmed words back into a single string
    return ' '.join(stems)

# Apply the stem_words function to 'tweet_content' column
corpus['tweet_content'] = corpus['tweet_content'].apply(stem_words)

# Display the modified DataFrame
print(corpus)
```
